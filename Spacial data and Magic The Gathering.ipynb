```python
% sql CREATE VIEW maxPopulation as SELECT * FROM geonames WHERE population=(SELECT MAX(population) FROM geonames);
```

```python
%sql SELECT * from maxPopulation;
```

```python
%sql SELECT name, (SELECT point(maxPopulation.latitude, maxPopulation.longitude) FROM maxPopulation) <-> point(geonames.latitude, geonames.longitude) as Distance FROM geonames ORDER BY Distance DESC LIMIT 100; 
```

```python
%%sql CREATE OR REPLACE FUNCTION public.geodistance(alat double precision, alng double precision, blat double precision, blng double precision)
  RETURNS double precision AS
$BODY$
SELECT asin(
  sqrt(
    sin(radians($3-$1)/2)^2 +
    sin(radians($4-$2)/2)^2 *
    cos(radians($1)) *
    cos(radians($3))
  )
) * 7926.3352 AS distance;
$BODY$
  LANGUAGE sql IMMUTABLE
COST 100;
```

```python
%sql SELECT geonames.name, public.geodistance(geonames.latitude, geonames.longitude, maxPopulation.latitude, maxPopulation.longitude) as Distance FROM geonames, maxPopulation ORDER BY Distance DESC LIMIT 100;
```

```python

%sql CREATE OR REPLACE VIEW distances as SELECT geonames.name, public.geodistance(geonames.latitude, geonames.longitude, maxPopulation.latitude, maxPopulation.longitude) as Distance, geonames.population FROM geonames, maxPopulation;
```

```python
%sql SELECT population, name, distance FROM distances ORDER BY population DESC LIMIT 10000;
```

```python
%%sql CREATE OR REPLACE VIEW maxThousand as
SELECT latitude, longitude, population FROM geonames ORDER BY population DESC LIMIT 1000;
```

```python
x = %sql SELECT latitude FROM maxThousand;
y = %sql SELECT longitude FROM maxThousand;
```
